```python
# import necessary libraries

import numpy as np
import cv2
import time
```

```python
# check the binaryzation of the image by the threshold value

# create a image that from white to black
img = np.zeros((100, 255), dtype=np.uint8)
for i in range(255):
    img[:, i] = i

# show the image before binaryzation
cv2.imshow("img", img)
cv2.waitKey(0)  # press any key to continue
cv2.destroyAllWindows() # close the window

# binaryzation
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)        
'''127 is the threshold value,
any value above 127 will be set to 255, 
any value below 127 will be set to 0'''


# print the threshold value and the image after binaryzation
print("ret: ", ret)
print("thresh: ", thresh)

# show the image
cv2.imshow("thresh", thresh)
cv2.waitKey(0)  # press any key to continue
cv2.destroyAllWindows() # close the window
```

```python
# check the thresh_binary_inv
ret, thresh_binary_inv = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
'''this is the inverse of the thresh_binary,
the value above 127 will be set to 0,
the value below 127 will be set to 255,
it is contrary to the thresh_binary that be shown on the last cell'''

# print the threshold value and the image after binaryzation
print("ret: ", ret)
print("thresh_binary_inv: ", thresh_binary_inv)

# show the image
cv2.imshow("thresh_binary_inv", thresh_binary_inv)
cv2.waitKey(0)  # press any key to continue
cv2.destroyAllWindows() # close the window
```

```python
# check the binaryization of thresh_tozero

# make the image from black to white
img = np.zeros((100, 255), dtype=np.uint8)
for i in range(255):
    img[:, i] = i

# binaryzation
ret, thresh_tozero = cv2.threshold(img, 127, 127, cv2.THRESH_TOZERO)
'''this is the binaryization of thresh_tozero,
the value above 127 will be not changed,
the value below 127 will be set to 0'''

# print the threshold value and the image after binaryzation
print("ret: ", ret)
print("thresh_tozero: ", thresh_tozero)

# show the image
cv2.imshow("thresh_tozero", thresh_tozero)
cv2.imshow("img", img)
cv2.waitKey(0)  # press any key to continue
cv2.destroyAllWindows() # close the window
```

```python
# check the binaryization of thresh_tozero_inv

# make the image from black to white
img = np.zeros((100, 255), dtype=np.uint8)
for i in range(255):
    img[:, i] = i

# binaryzation
ret, thresh_tozero_inv = cv2.threshold(img, 127, 127, cv2.THRESH_TOZERO_INV)
'''this is the inverse of the binaryization of thresh_tozero,
the value above 127 will be set to 0,
the value below 127 will be not changed'''

# print the threshold value and the image after binaryzation
print("ret: ", ret)
print("thresh_tozero_inv: ", thresh_tozero_inv)

# show the image
cv2.imshow("thresh_tozero_inv", thresh_tozero_inv)
cv2.imshow("img", img)
cv2.waitKey(0)  # press any key to continue
cv2.destroyAllWindows() # close the window
```

```python
# check the binaryization of thresh_trunc

# make the image from black to white
img = np.zeros((100, 255), dtype=np.uint8)
for i in range(255):
    img[:, i] = i

# binaryzation
ret, thresh_trunc = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
'''this is the binaryization of thresh_trunc,
the value above 127 will be set to the threshold value,
the value below 127 will be not changed'''

# print the threshold value and the image after binaryzation
print("ret: ", ret)
print("thresh_trunc: ", thresh_trunc)

# show the image
cv2.imshow("thresh_trunc", thresh_trunc)
cv2.imshow("img", img)
cv2.waitKey(0)  # press any key to continue
cv2.destroyAllWindows() # close the window
```

```python
# show the different binaryization methods

# read the image
img = cv2.imread("lena.png")
# convert the image to gray
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

t1, thresh_binary = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
t2, thresh_binary_inv = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)
t3, thresh_trunc = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TRUNC)
t4, thresh_tozero = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO)
t5, thresh_tozero_inv = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO_INV)

# show the image
cv2.imshow("thresh_binary", thresh_binary)
cv2.imshow("thresh_binary_inv", thresh_binary_inv)
cv2.imshow("thresh_trunc", thresh_trunc)
cv2.imshow("thresh_tozero", thresh_tozero)
cv2.imshow("thresh_tozero_inv", thresh_tozero_inv)
cv2.waitKey(0)  # press any key to continue
cv2.destroyAllWindows() # close the window
```

```python
# self-adaptive binaryzation

# read the image
img = cv2.imread("lena.png")
# convert the image to gray
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# compute the threshold value by cv2.adaptive_thresh_mean_c
thresh_mean_c = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 5, 7)

# compute the threshold value by cv2.adaptive_thresh_gaussian_c
thresh_gaussian_c = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 5, 7)

# show the image
cv2.imshow("thresh_mean_c", thresh_mean_c)
cv2.imshow("thresh_gaussian_c", thresh_gaussian_c)
cv2.waitKey(0)  # press any key to continue
cv2.destroyAllWindows() # close the window
```

```python
# check binaryization by otsu method

# compute the running time of the binaryzation by otsu method
start = time.time()

# read the image
img = cv2.imread("lena.png")
# convert the image to gray
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# normal binaryzation
ret, thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
# binaryzation by otsu method
ret, thresh_otsu = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

end = time.time()
print("time: ", end - start)

# show the image
cv2.imshow("thresh", thresh)
cv2.imshow("thresh_otsu", thresh_otsu)
cv2.waitKey(0)  # press any key to continue
cv2.destroyAllWindows() # close the window
```

```python
# practice: binaryzation of the image

# read the image
img = cv2.imread("car.jpg")
# convert the image to gray
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# binaryzation
ret, thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)  # normal binaryzation
ret, thresh_inv = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)  # inverse of thresh
ret, thresh_otsu = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)  # binaryzation by otsu method

# show four images by using plt.subplot
import matplotlib.pyplot as plt
# turn the original image to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# set the size of the figure
plt.figure(figsize=(10, 10))
# show the original image
plt.subplot(2, 3, 1)
plt.title("original image")
plt.imshow(img_rgb)
# show the normal binaryzation
plt.subplot(2, 3, 2)
plt.title("normal binaryzation")
plt.imshow(thresh, cmap="gray")
# show the inverse of thresh
plt.subplot(2, 3, 3)
plt.title("inverse of thresh")
plt.imshow(thresh_inv, cmap="gray")
# show the binaryzation by otsu method
plt.subplot(2, 3, 4)
plt.title("binaryzation by otsu method")
plt.imshow(thresh_otsu, cmap="gray")

# show the image
plt.show()


```
